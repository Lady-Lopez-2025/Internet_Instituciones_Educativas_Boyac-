```python
#Importamos las librerias que vamos a utilizar

import pandas as pf
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Guardo los datos del CSV en un Data Frame llamado df
df=pf.read_csv("IED_Boyaca_Internet_clean.csv")
```

```python
#Analizaremos las provincias del Departamento de Boyacá que cuentan con servicio de internet en las instituciones educativas

df.groupby('PROVINCIA')['ESTADO'].value_counts().reset_index()

```

```python
#Analizamos las provincias del departamento de Boyacá y el estado del servicio de internet

conteo = df.groupby(['PROVINCIA', 'ESTADO']).size().reset_index(name='Total')


plt.figure(figsize=(12, 6))
sns.barplot(data=conteo, x='PROVINCIA', y='Total', hue='ESTADO', palette='Set2')

plt.title('Cantidad de instituciones por provincia y estado del servicio')
plt.xlabel('PROVINCIA')
plt.ylabel('ESTADO DEL SERVICIO')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
#Con estos datos podemos analizar que la provincia del occidente es la más critica ya que se concentra en esta la mayor cantidad de instituciones sin servicio de internet.
#Mientras que las provincias de Centro y Sugamuxi están mejor "equilibradas" en cuando a instituciones con servicio de internet o en periodo de transicion a contar con el servicio.
```

```python
#Analizamos las provincias del departamento de Boyacá y las zonas que cuentan con el servicio de internet

conteo = df.groupby(['PROVINCIA', 'ZONA']).size().reset_index(name='Total')


plt.figure(figsize=(12, 6))
sns.barplot(data=conteo, x='PROVINCIA', y='Total', hue='ZONA', palette='Set2')

plt.title('Cantidad de instituciones por provincia y zona')
plt.xlabel('PROVINCIA')
plt.ylabel('ZONA')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
